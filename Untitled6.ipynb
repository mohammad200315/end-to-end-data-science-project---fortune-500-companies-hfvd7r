```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.io as pio
```

```python
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', '{:,.2f}'.format)

class Fortune500Analyzer:
    def __init__(self, data_path):
        self.df = pd.read_csv(data_path)
        self._validate_data()
        self._create_features()

    def _validate_data(self):
        print("Data Validation Report")
        print("="*50)

        validation_report = {
            'Total Rows': len(self.df),
            'Total Columns': len(self.df.columns),
            'Missing Values': self.df.isnull().sum().sum(),
            'Duplicate Rows': self.df.duplicated().sum(),
            'Memory Usage': f"{self.df.memory_usage(deep=True).sum() / 1024**2:.2f} MB"
        }

        for key, value in validation_report.items():
            print(f"{key}: {value}")

        print("\nData Types Distribution:")
        print(self.df.dtypes.value_counts())

    def _create_features(self):
        self.df['profit_margin'] = (self.df['profit_mil'] / self.df['revenue_mil']) * 100
        self.df['revenue_per_employee'] = self.df['revenue_mil'] / self.df['employees']
        self.df['years_in_list'] = self.df['year'] - 1995
        self.df['size_category'] = pd.qcut(self.df['revenue_mil'], q=4, labels=['Small', 'Medium', 'Large', 'Giant'])
        self.df['decade'] = (self.df['year'] // 10) * 10

    def time_series_analysis(self):
        print("\nAdvanced Time Series Analysis")
        print("="*50)

        fig = make_subplots(
            rows=2, cols=2,
            subplot_titles=('Revenue Trend', 'Profit Margin Trend',
                          'Top 5 Companies Evolution', 'Industry Concentration'),
            specs=[[{'type': 'scatter'}, {'type': 'scatter'}],
                   [{'type': 'scatter'}, {'type': 'bar'}]]
        )

        yearly_stats = self.df.groupby('year').agg({
            'revenue_mil': 'mean',
            'profit_margin': 'mean',
            'name': 'count'
        }).rename(columns={'name': 'company_count'})

        fig.add_trace(
            go.Scatter(x=yearly_stats.index, y=yearly_stats['revenue_mil'],
                      mode='lines+markers', name='Avg Revenue',
                      line=dict(color='blue', width=3)),
            row=1, col=1
        )

        fig.add_trace(
            go.Scatter(x=yearly_stats.index, y=yearly_stats['profit_margin'],
                      mode='lines+markers', name='Profit Margin %',
                      line=dict(color='green', width=3)),
            row=1, col=2
        )

        top_companies = self.df.groupby('name')['revenue_mil'].max().nlargest(5).index
        for company in top_companies:
            company_data = self.df[self.df['name'] == company].sort_values('year')
            fig.add_trace(
                go.Scatter(x=company_data['year'], y=company_data['revenue_mil'],
                          mode='lines', name=company[:20]),
                row=2, col=1
            )

        industry_decade = pd.crosstab(self.df['decade'], self.df['industry']).sum().nlargest(10)
        fig.add_trace(
            go.Bar(x=industry_decade.index, y=industry_decade.values,
                   marker_color='coral'),
            row=2, col=2
        )

        fig.update_layout(height=800, showlegend=True, title_text="Fortune 500 Time Series Analysis")
        fig.show()

    def financial_ratio_analysis(self):
        print("\nAdvanced Financial Ratio Analysis")
        print("="*50)

        financial_metrics = self.df.groupby('year').agg({
            'revenue_mil': ['mean', 'std', lambda x: x.skew()],
            'profit_margin': ['mean', 'std'],
            'revenue_per_employee': 'mean',
            'asset_mil': lambda x: (self.df['revenue_mil'] / self.df['asset_mil']).mean()
        })

        financial_metrics.columns = ['Avg Revenue', 'Revenue Std', 'Revenue Skew',
                                   'Avg Profit Margin', 'Profit Margin Std',
                                   'Revenue/Employee', 'Asset Turnover']

        print(financial_metrics.round(3))

        numeric_df = self.df.select_dtypes(include=[np.number])
        correlation_matrix = numeric_df.corr()

        plt.figure(figsize=(12, 10))
        mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
        sns.heatmap(correlation_matrix, mask=mask, annot=True, fmt='.2f',
                   cmap='coolwarm', center=0, square=True, linewidths=.5,
                   cbar_kws={"shrink": .8})
        plt.title('Financial Metrics Correlation Heatmap', fontsize=16, pad=20)
        plt.tight_layout()
        plt.show()

    def industry_cluster_analysis(self):
        print("\nIndustry Cluster Analysis")
        print("="*50)

        industry_stats = self.df.groupby('industry').agg({
            'revenue_mil': ['count', 'mean', 'median', 'std'],
            'profit_margin': 'mean',
            'market_value_mil': 'median'
        }).round(2)

        industry_stats.columns = ['Count', 'Avg Revenue', 'Median Revenue',
                                 'Revenue Std', 'Avg Profit Margin', 'Median Market Value']

        industry_stats = industry_stats.sort_values('Avg Revenue', ascending=False)

        print("Top 10 Industries by Average Revenue:")
        print(industry_stats.head(10))

        fig = px.scatter(industry_stats.reset_index(),
                        x='Avg Revenue',
                        y='Avg Profit Margin',
                        size='Count',
                        color='Median Market Value',
                        hover_name='industry',
                        log_x=True,
                        size_max=60,
                        title='Industry Performance Clusters',
                        labels={'Avg Revenue': 'Average Revenue (Millions USD)',
                               'Avg Profit Margin': 'Average Profit Margin (%)',
                               'Count': 'Number of Companies',
                               'Median Market Value': 'Median Market Value (Millions USD)'})

        fig.update_layout(height=600)
        fig.show()

    def geospatial_analysis(self):
        print("\nAdvanced Geospatial Analysis")
        print("="*50)

        if 'headquarters_state' in self.df.columns:
            state_analysis = self.df.groupby('headquarters_state').agg({
                'name': 'count',
                'revenue_mil': 'sum',
                'profit_mil': 'sum',
                'employees': 'sum'
            }).sort_values('revenue_mil', ascending=False)

            state_analysis.columns = ['Company Count', 'Total Revenue', 'Total Profit', 'Total Employees']

            print("Top 10 States by Total Revenue:")
            print(state_analysis.head(10))

            try:
                fig = px.choropleth(state_analysis.reset_index(),
                                   locations='headquarters_state',
                                   locationmode="USA-states",
                                   color='Total Revenue',
                                   scope="usa",
                                   color_continuous_scale="Viridis",
                                   title='Fortune 500 Revenue Distribution by State',
                                   hover_data=['Company Count', 'Total Profit'])
                fig.show()
            except:
                plt.figure(figsize=(12, 6))
                state_analysis.head(15).plot(kind='bar', y='Total Revenue',
                                           title='Top 15 States by Total Revenue')
                plt.xticks(rotation=45)
                plt.tight_layout()
                plt.show()

    def predictive_insights(self):
        print("\nPredictive Insights & Trends")
        print("="*50)

        from sklearn.linear_model import LinearRegression

        yearly_avg = self.df.groupby('year')['revenue_mil'].mean().reset_index()
        X = yearly_avg[['year']]
        y = yearly_avg['revenue_mil']

        model = LinearRegression()
        model.fit(X, y)

        print(f"Revenue Growth Trend: {model.coef_[0]:,.0f} per year")
        print(f"R² Score: {model.score(X, y):.3f}")

        future_years = np.array([[2024], [2025], [2026], [2027], [2028]])
        predictions = model.predict(future_years)

        print("\nRevenue Predictions (Millions USD):")
        for year, pred in zip([2024, 2025, 2026, 2027, 2028], predictions):
            print(f"  {year}: ${pred:,.0f}")

    def generate_report(self):
        print("\nGenerating Comprehensive Analysis Report...")
        print("="*60)

        report = {
            'Data Summary': {
                'Period': f"{self.df['year'].min()} - {self.df['year'].max()}",
                'Total Observations': len(self.df),
                'Unique Companies': self.df['name'].nunique(),
                'Industries Covered': self.df['industry'].nunique()
            },
            'Financial Highlights': {
                'Total Revenue (All Years)': f"${self.df['revenue_mil'].sum():,.0f}M",
                'Average Revenue': f"${self.df['revenue_mil'].mean():,.0f}M",
                'Highest Revenue': f"${self.df['revenue_mil'].max():,.0f}M",
                'Average Profit Margin': f"{self.df['profit_margin'].mean():.1f}%"
            },
            'Trend Analysis': {
                'Revenue CAGR': self._calculate_cagr(),
                'Most Frequent Industry': self.df['industry'].mode()[0],
                'Top State': self.df['headquarters_state'].mode()[0],
                'Average Company Tenure': f"{self.df['years_in_list'].mean():.1f} years"
            }
        }

        print("\nEXECUTIVE SUMMARY")
        print("-"*40)
        for section, metrics in report.items():
            print(f"\n{section}:")
            for metric, value in metrics.items():
                print(f"  • {metric}: {value}")

    def _calculate_cagr(self):
        first_year = self.df['year'].min()
        last_year = self.df['year'].max()

        first_avg = self.df[self.df['year'] == first_year]['revenue_mil'].mean()
        last_avg = self.df[self.df['year'] == last_year]['revenue_mil'].mean()

        years = last_year - first_year
        cagr = ((last_avg / first_avg) ** (1/years) - 1) * 100

        return f"{cagr:.1f}%"

if __name__ == "__main__":
    analyzer = Fortune500Analyzer('/content/fortune500_cleaned.csv')

    analyzer.generate_report()
    analyzer.time_series_analysis()
    analyzer.financial_ratio_analysis()
    analyzer.industry_cluster_analysis()
    analyzer.geospatial_analysis()
    analyzer.predictive_insights()
```
